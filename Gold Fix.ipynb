```python
import pandas as pd

# Load both CSVs (DS file uses semicolons)
gold = pd.read_csv('/Users/marc/Desktop/Gold Futures Historical Datacsv.csv')
ds = pd.read_csv('/Users/marc/Desktop/DS project M2 & Asset Prices - Group D(Variables).csv', sep=";")

# Convert 'Date' columns to datetime
gold["Date"] = pd.to_datetime(gold["Date"], errors="coerce")
ds["Date"] = pd.to_datetime(ds["Date"], errors="coerce")

# Drop invalid or missing dates
gold = gold.dropna(subset=["Date"])
ds = ds.dropna(subset=["Date"])

# Find extra dates in Gold not in DS
extra_dates = sorted(set(gold["Date"]) - set(ds["Date"]))

# Display result
print("Extra dates in Gold Futures (not in DS Variables):")
for d in extra_dates:
    print(d.date())

print(f"\nTotal extra dates found: {len(extra_dates)}")
```
