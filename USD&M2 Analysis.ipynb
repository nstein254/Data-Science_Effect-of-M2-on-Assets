```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# -----------------------
# Configuration
# -----------------------

# Path to the CSV file containing percentage changes
FILE = Path("/Users/babo/Downloads/Group-project-Data_Percent_Changes.csv")

# Column names in the dataset
date_col = "Date"
m2_col = "M2 Money Supply (Bn)"     # M2 (% change)
usd_col = "US Dollar Index"         # USD Index (% change)

# Define the financial crisis time window
start_date = "2008-01-01"
end_date   = "2009-06-30"

# Define the core crisis (Lehman collapse period)
shock_start = "2008-09-07"
shock_end   = "2009-03-22"

# -----------------------
# Load and prepare data
# -----------------------

# Load the CSV file
df = pd.read_csv(FILE)

# Convert Date column to datetime format
df[date_col] = pd.to_datetime(df[date_col], errors="coerce")

# Sort data by date so time series plots are correct
df = df.sort_values(date_col)

# Convert M2 and USD Index columns to numeric values
df[m2_col] = pd.to_numeric(df[m2_col], errors="coerce")
df[usd_col] = pd.to_numeric(df[usd_col], errors="coerce")

# Filter the dataset to the selected financial crisis window
mask = (df[date_col] >= start_date) & (df[date_col] <= end_date)
df_win = df.loc[mask, [date_col, m2_col, usd_col]].dropna()

# -----------------------
# Dual y-axis plot
# -----------------------

# Create the figure and first y-axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot USD Index (% change) on the left axis
ax1.plot(
    df_win[date_col],
    df_win[usd_col],
    color="navy",
    linewidth=2,
    label="USD Index (% change)"
)
ax1.set_ylabel("USD Index (% change)", color="navy")
ax1.tick_params(axis="y", labelcolor="navy")

# Create second y-axis for M2
ax2 = ax1.twinx()

# Plot M2 growth (% change) on the right axis
ax2.plot(
    df_win[date_col],
    df_win[m2_col],
    color="green",
    linewidth=2,
    label="M2 (% change)"
)
ax2.set_ylabel("M2 (% change)", color="green")
ax2.tick_params(axis="y", labelcolor="green")

# Highlight the core financial crisis period
ax1.axvspan(
    pd.to_datetime(shock_start),
    pd.to_datetime(shock_end),
    alpha=0.2,
    color="grey",
    label="Financial crisis core"
)

# Add title
plt.title("Financial Crisis (20082009): USD Index vs M2 Growth")

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left")

# Improve readability
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
